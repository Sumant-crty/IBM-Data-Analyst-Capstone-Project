#### Step 2: Load and Preview the Dataset
Load the dataset from the provided URL. Use df.head() to display the first few rows to get an overview of the structure.


```python
# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)

# Display the first few rows of the dataset
df.head()
```

#### Step 3: Handling Missing Data


Identify and manage missing values in critical columns such as `Employment`, `JobSat`, and `RemoteWork`. Implement a strategy to fill or drop these values, depending on the significance of the missing data.


```python
print('Missing values before handling:')
print(df[['Employment', 'JobSat', 'RemoteWork']].isnull().sum())

# For 'Employment', fill NaN with 'Not Employed' or similar if reasonable, or drop.
# Given it's a key demographic, dropping might be better if the count is small,
# or if 'Student' etc. are also considered 'Employment'.
# Let's check unique values first.
print('\nUnique values in Employment before handling NaN:')
print(df['Employment'].value_counts(dropna=False))

# For 'RemoteWork', fill NaN with 'No info' or 'Not applicable' or assume 'None' if appropriate.
# Let's assume 'None' for now, indicating they are not working remotely if not specified.
print('\nUnique values in RemoteWork before handling NaN:')
print(df['RemoteWork'].value_counts(dropna=False))

# For 'JobSat', filling with a neutral value like the median/mode might skew results.
# It's often better to drop rows where JobSat is NaN for satisfaction analysis.
print('\nUnique values in JobSat before handling NaN:')
print(df['JobSat'].value_counts(dropna=False))

# Strategy:
# 1. For 'Employment', if NaN represents non-response, we might keep it or fill with 'Unknown'.
#    For now, let's keep NaNs for employment type as they might be distinct.
# 2. For 'RemoteWork', filling with 'Not applicable' seems reasonable for those not working.
#    Or we can fill with 'No response' to distinguish from 'None'.
#    Let's fill with 'No response' for now.
df['RemoteWork'].fillna('No response', inplace=True)

# 3. For 'JobSat', dropping rows with NaN is generally the safest approach for satisfaction analysis
#    to avoid introducing bias. Since JobSat is critical for subsequent analysis,
#    we will drop rows where JobSat is NaN.
df.dropna(subset=['JobSat'], inplace=True)

# Recheck missing values after handling
print('\nMissing values after handling:')
print(df[['Employment', 'JobSat', 'RemoteWork']].isnull().sum())

# Display the first few rows to see the changes
display(df.head())
```

#### Step 4: Analysis of Experience and Job Satisfaction


Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
# Convert 'YearsCodePro' to numeric, coercing errors to NaN
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop rows where 'YearsCodePro' is NaN for this analysis to ensure valid ranges
df_exp = df.dropna(subset=['YearsCodePro']).copy()

# Create experience ranges
bins = [-1, 5, 10, 20, df_exp['YearsCodePro'].max()]
labels = ['0-5 years', '6-10 years', '11-20 years', '>20 years']
df_exp['ExperienceRange'] = pd.cut(df_exp['YearsCodePro'], bins=bins, labels=labels, right=True)

# Calculate median JobSat for each experience range, explicitly setting observed=False
median_job_sat = df_exp.groupby('ExperienceRange', observed=False)['JobSat'].median().reset_index()

print("Median Job Satisfaction by Experience Range:")
print(median_job_sat)

# Visualize the relationship, addressing the FutureWarning by assigning hue and setting legend=False
plt.figure(figsize=(10, 6))
sns.barplot(x='ExperienceRange', y='JobSat', data=median_job_sat, palette='viridis', hue='ExperienceRange', legend=False)
plt.title('Median Job Satisfaction by Professional Experience Range')
plt.xlabel('Professional Experience Range')
plt.ylabel('Median Job Satisfaction')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Step 5: Visualize Job Satisfaction


Use a count plot to show the distribution of `JobSat` values. This provides insights into the overall satisfaction levels of respondents.


```python
# Use a count plot to show the distribution of JobSat values
plt.figure(figsize=(10, 6))
sns.countplot(x='JobSat', data=df, palette='viridis', hue='JobSat', legend=False)
plt.title('Distribution of Job Satisfaction')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Number of Respondents')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Step 6: Analyzing Remote Work Preferences by Job Role


Analyze trends in remote work based on job roles. Use the `RemoteWork` and `Employment` columns to explore preferences and examine if specific job roles prefer remote work more than others.

- Use a count plot to show remote work distribution.

- Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.


```python
# Use a count plot to show remote work distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='RemoteWork', data=df, palette='viridis', hue='RemoteWork', legend=False)
plt.title('Distribution of Remote Work Preferences')
plt.xlabel('Remote Work Type')
plt.ylabel('Number of Respondents')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Cross-tabulate remote work preferences by employment type
# This will show how many of each employment type prefer each remote work option
employment_remote_crosstab = pd.crosstab(df['Employment'], df['RemoteWork'])
print("\nCross-tabulation of Remote Work by Employment Type:")
print(employment_remote_crosstab)

# For job roles (DevType), we need to handle multiple selections.
# First, filter out NaN values in DevType
df_devtype = df.dropna(subset=['DevType']).copy()

# Split the 'DevType' column into individual roles and explode the DataFrame
df_devtype['DevType_split'] = df_devtype['DevType'].str.split(';')
df_devtype_exploded = df_devtype.explode('DevType_split')

# Strip whitespace from the split roles
df_devtype_exploded['DevType_split'] = df_devtype_exploded['DevType_split'].str.strip()

# Now, cross-tabulate remote work preferences by individual job roles
devtype_remote_crosstab = pd.crosstab(df_devtype_exploded['DevType_split'], df_devtype_exploded['RemoteWork'])
print("\nCross-tabulation of Remote Work by Job Roles:")
print(devtype_remote_crosstab)

# Optional: Visualize top N job roles' remote work preferences
# Get the top N job roles by total count
top_n_devtypes = df_devtype_exploded['DevType_split'].value_counts().head(10).index
devtype_remote_crosstab_top = devtype_remote_crosstab.loc[top_n_devtypes]

devtype_remote_crosstab_top.plot(kind='bar', figsize=(14, 7), stacked=True)
plt.title('Remote Work Preferences by Top 10 Job Roles (Stacked Bar Plot)')
plt.xlabel('Job Role')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Remote Work')
plt.tight_layout()
plt.show()
```

#### Step 7: Analyzing Programming Language Trends by Region


Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.


```python
# First, let's clean the 'Country' column by stripping whitespace
df['Country'] = df['Country'].str.strip()

# Filter out rows where 'LanguageHaveWorkedWith' is NaN
df_languages = df.dropna(subset=['LanguageHaveWorkedWith']).copy()

# Split the 'LanguageHaveWorkedWith' column into individual languages and explode the DataFrame
df_languages['Language_split'] = df_languages['LanguageHaveWorkedWith'].str.split(';')
df_languages_exploded = df_languages.explode('Language_split')

# Strip whitespace from the split languages
df_languages_exploded['Language_split'] = df_languages_exploded['Language_split'].str.strip()

# Get the top N countries to focus the analysis for better visualization
top_n_countries = df_languages_exploded['Country'].value_counts().head(10).index
df_top_countries = df_languages_exploded[df_languages_exploded['Country'].isin(top_n_countries)]

# Calculate the count of each language per country
language_country_crosstab = pd.crosstab(df_top_countries['Country'], df_top_countries['Language_split'])

# Get the top 10 most popular languages overall for better visualization
top_10_languages = df_languages_exploded['Language_split'].value_counts().head(10).index
language_country_crosstab_top_10 = language_country_crosstab[top_10_languages]

print("Top 10 Programming Languages by Top 10 Countries:")
print(language_country_crosstab_top_10)

# Visualize the top programming languages by region (top 10 countries and top 10 languages)
plt.figure(figsize=(16, 10))
sns.heatmap(language_country_crosstab_top_10, annot=True, fmt="d", cmap="viridis", linewidths=.5)
plt.title('Top 10 Programming Languages by Top 10 Countries')
plt.xlabel('Programming Language')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

#### Step 8: Correlation Between Experience and Satisfaction


Examine how years of experience (`YearsCodePro`) correlate with job satisfaction (`JobSatPoints_1`). Use a scatter plot to visualize this relationship.


```python
# To examine the correlation between YearsCodePro and JobSat, we first need to ensure both are numeric.
# YearsCodePro was already converted to numeric and NaNs dropped in Step 4 for df_exp.
# JobSat is already numeric after handling NaNs in Step 3.

# Let's use the df_exp DataFrame from Step 4 as it already has YearsCodePro as numeric and handles NaNs appropriately for this analysis.
# Filter out rows where YearsCodePro is NaN if we didn't already for df_exp
df_correlation = df.dropna(subset=['YearsCodePro', 'JobSat']).copy()

# Convert 'YearsCodePro' to numeric, coercing errors to NaN
df_correlation['YearsCodePro'] = pd.to_numeric(df_correlation['YearsCodePro'], errors='coerce')

# Drop rows where 'YearsCodePro' is NaN after conversion if any non-numeric values were coerced
df_correlation.dropna(subset=['YearsCodePro'], inplace=True)

# Calculate the correlation coefficient
correlation = df_correlation['YearsCodePro'].corr(df_correlation['JobSat'])
print(f"Correlation between YearsCodePro and JobSat: {correlation:.2f}")

# Create a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsCodePro', y='JobSat', data=df_correlation, alpha=0.5)
plt.title('Correlation between Professional Experience and Job Satisfaction')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Job Satisfaction Score')
plt.grid(True)
plt.show()
```

#### Step 9: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
# Cross-tabulate educational background and employment type
ed_level_employment_crosstab = pd.crosstab(df['EdLevel'], df['Employment'])

print("Cross-tabulation of Educational Background by Employment Type:")
print(ed_level_employment_crosstab)

# Visualize the relationship (e.g., stacked bar plot for top employment types by education)
# First, let's simplify 'Employment' to a few main categories for better visualization, if it has too many unique values.
# Check value counts for Employment to decide on simplification
print('\nUnique values and counts for Employment:')
print(df['Employment'].value_counts())

# For visualization, let's pick a few prominent employment types or aggregate them.
# Let's consider 'Employed, full-time', 'Student, full-time', 'Independent contractor, freelancer, or self-employed'
# and others will be grouped as 'Other'.

def categorize_employment(employment_str):
    if 'Employed, full-time' in employment_str:
        return 'Employed, Full-time'
    elif 'Student, full-time' in employment_str:
        return 'Student, Full-time'
    elif 'Independent contractor' in employment_str or 'freelancer' in employment_str or 'self-employed' in employment_str:
        return 'Freelancer/Contractor'
    else:
        return 'Other'

df_viz = df.copy()
df_viz['Employment_Category'] = df_viz['Employment'].apply(categorize_employment)

# Create a simplified crosstab for visualization
ed_employment_viz = pd.crosstab(df_viz['EdLevel'], df_viz['Employment_Category'])

# Normalize the data to show proportions for better comparison across educational levels
ed_employment_viz_normalized = ed_employment_viz.divide(ed_employment_viz.sum(axis=1), axis=0)

plt.figure(figsize=(14, 8))
ed_employment_viz_normalized.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Proportion of Employment Types by Educational Level')
plt.xlabel('Educational Level')
plt.ylabel('Proportion of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Type Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
