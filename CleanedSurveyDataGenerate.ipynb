```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("survey_data_updated 5.csv")

# Split languages (assuming they are separated by ';')
languages = df['LanguageHaveWorkedWith'].dropna().str.split(';')

# Flatten the list
all_languages = [lang for sublist in languages for lang in sublist]

# Count frequency
lang_counts = pd.Series(all_languages).value_counts().head(10)

# Plot horizontal bar chart
plt.figure()
lang_counts.plot(kind='barh')

plt.title('Top 10 Languages  Have Worked With')
plt.ylabel('Programming Languages')
plt.xlabel('Count')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Step 1: Load dataset
df = pd.read_csv("survey_data_updated 5.csv")

# Step 2: Split platforms (separated by ;)
platforms = df['PlatformHaveWorkedWith'].dropna().str.split(';')

# Step 3: Flatten list
all_platforms = [p for sublist in platforms for p in sublist]

# Step 4: Count frequency and take Top 10
top10 = pd.Series(all_platforms).value_counts().head(10)

# Step 5: Convert to dictionary (required for WordCloud)
freq_dict = top10.to_dict()

# Step 6: Generate Word Cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white'
).generate_from_frequencies(freq_dict)

# Step 7: Display
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

plt.title('Top 10 Platforms Worked With (Word Cloud)')
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("survey_data_updated 5.csv")

# Step 2: Split frameworks (separated by ;)
wf_split = df['WebframeHaveWorkedWith'].dropna().str.split(';')

# Step 3: Flatten list
all_wf = [wf for sublist in wf_split for wf in sublist]

# Step 4: Count frequency and get Top 10
top10_wf = pd.Series(all_wf).value_counts().head(10)

# Step 5: Prepare data
labels = top10_wf.index
sizes = top10_wf.values

# Create positions for bubbles
x = range(len(labels))
y = [1] * len(labels)  # keep on same line (like hierarchy row)

# Step 6: Plot bubble chart
plt.figure()

scatter = plt.scatter(
    x, y,
    s=[s * 10 for s in sizes],   # bubble size scaling
    c=sizes,                     # color based on frequency
)

# Step 7: Add labels
for i, label in enumerate(labels):
    plt.text(x[i], y[i], label, ha='center', va='center')

# Step 8: Styling
plt.title('Top 10 Web Frameworks Worked With (Bubble Chart)')
plt.axis('off')  # hide axes for clean look
plt.colorbar(scatter, label='Count')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Step 1: Load dataset
df = pd.read_csv("survey_data_updated 5.csv")

# Step 2: Columns to clean (fixing typos as well)
columns_to_clean = [
    'LanguageHaveWorkedWith',
    'LanguageWantToWorkWith',
    'DatabaseHaveWorkedWith',
    'DatabaseWantToWorkWith',
    'PlatformHaveWorkedWith',
    'PlatformWantToWorkWith',
    'WebframeHaveWorkedWith',
    'WebframeWantToWorkWith'
]

# Step 3: Cleaning function
def clean_multivalue_column(col):
    return (
        col.fillna('')  # replace NaN
           .apply(lambda x: ';'.join(
               sorted(set(  # remove duplicates + sort
                   [i.strip() for i in x.split(';') if i.strip() != '']
               ))
           ))
    )

# Step 4: Apply cleaning
for col in columns_to_clean:
    if col in df.columns:
        df[col] = clean_multivalue_column(df[col])

# Step 5: Optional - remove leading/trailing spaces in ALL columns
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Step 6: Save cleaned dataset
df.to_csv("cleaned_survey_data.csv", index=False)

print("✅ Cleaned file saved as: cleaned_survey_data.csv")
```
