### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
print(df.info())
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
print(df.describe())
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Identify inconsistent or irrelevant entries in 'Country'
print("Unique values and their counts in 'Country' column:")
print(df['Country'].value_counts())

# Identify inconsistent or irrelevant entries in 'EdLevel'
print("\nUnique values and their counts in 'EdLevel' column:")
print(df['EdLevel'].value_counts())
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
# Standardize entries in 'Country' column (example: handling common variations)
country_mapping = {
    'United States of America': 'USA',
    'United Kingdom': 'UK',
    'Germany': 'Germany',
    # Add more mappings as needed
}
df['Country'] = df['Country'].replace(country_mapping)
print("Standardized 'Country' values (first 10 unique values):\n", df['Country'].value_counts().head(10))

# Standardize entries in 'EdLevel' column (example: consolidating similar entries)
edlevel_mapping = {
    'Primary/elementary school': 'Less than a Bachelor\'s Degree',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'Less than a Bachelor\'s Degree',
    'Some college/university study without earning a degree': 'Less than a Bachelor\'s Degree',
    'Associate degree (A.A., A.S., etc.)': 'Less than a Bachelor\'s Degree',
    "Bachelor's degree (B.A., B.S., B.Eng., etc.)": "Bachelor's Degree",
    "Master's degree (M.A., M.S., M.Eng., MBA, etc.)": "Master's Degree",
    'Other doctoral degree (Ph.D., Ed.D., etc.)': 'Doctoral Degree',
    'Professional degree (JD, MD, etc.)': 'Doctoral Degree'
}
df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)
print("\nStandardized 'EdLevel' values:\n", df['EdLevel'].value_counts())
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
# Perform one-hot encoding on the 'Employment' column
df_encoded = pd.get_dummies(df, columns=['Employment'], prefix='Employment')

# Display the first few rows of the DataFrame with the new encoded columns
print(df_encoded.head())
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
# Calculate the number of missing values for each column
missing_values = df.isnull().sum()

# Sort the columns by the number of missing values in descending order
missing_values_sorted = missing_values.sort_values(ascending=False)

# Display the columns with the highest number of missing values
print("Columns with the highest number of missing values:\n", missing_values_sorted.head(10))
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
# Impute missing values in 'ConvertedCompYearly' with the median
median_converted_comp = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'].fillna(median_converted_comp, inplace=True)

# Verify that missing values have been handled
print("Missing values in 'ConvertedCompYearly' after imputation:", df['ConvertedCompYearly'].isnull().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
# Impute missing values in 'RemoteWork' with the most frequent value (mode)
mode_remotework = df['RemoteWork'].mode()[0] # .mode() can return multiple values, so take the first one
df['RemoteWork'].fillna(mode_remotework, inplace=True)

# Verify that missing values have been handled
print("Missing values in 'RemoteWork' after imputation:", df['RemoteWork'].isnull().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max Scaling to 'ConvertedCompYearly'
# Reshape the column as it expects a 2D array
df['ConvertedCompYearly_Scaled'] = scaler.fit_transform(df[['ConvertedCompYearly']])

# Display the first few rows with the new scaled column
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Scaled']].head())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
import numpy as np

# Apply log transformation to 'ConvertedCompYearly'
# Use np.log1p to handle potential zero values by adding 1 before log transformation
df['ConvertedCompYearly_Log'] = np.log1p(df['ConvertedCompYearly'])

# Display the first few rows with the new log-transformed column
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
# Define bins for experience levels
bins = [0, 5, 15, 30, np.inf]
labels = ['Entry-level', 'Mid-level', 'Senior-level', 'Expert']

# Create the 'ExperienceLevel' column based on 'YearsCodePro'
# First, convert 'YearsCodePro' to numeric, coercing errors to NaN
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Fill NaN values in 'YearsCodePro' with a value that puts them in a reasonable category, or handle separately
# For simplicity, let's fill with 0 to put them in 'Entry-level' if NaN implies no professional experience
df['YearsCodePro'].fillna(0, inplace=True)

df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], bins=bins, labels=labels, right=False)

# Display the counts of the new 'ExperienceLevel' column
print(df['ExperienceLevel'].value_counts())
print(df[['YearsCodePro', 'ExperienceLevel']].head())
```
